# 다중 서브 플롯

subplots() 함수 
: 하나의 큰 그림 내부에 있는 작은 그림



```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
nrows,nclos=3,4
```

plt.subplot._adjust 함수를 이용하여 높이 와 푹의 크기를 설정

```python
fig = plt.figure()
fig.subplots_adjust
(hspace=0.4,wspace=0.4)
```

plt.subplots 함수 를 사용하연 간단하게 표현가능

```python
nrows,ncols=3,4
fig,ax = plt.subplots(nrows,ncols,sharex='col', sharey='row')
```
